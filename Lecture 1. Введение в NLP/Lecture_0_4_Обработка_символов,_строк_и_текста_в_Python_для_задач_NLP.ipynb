Давайте разберем основные функции и методы для работы с символами, строками и текстом в Python, особенно в контексте обработки естественного языка (Natural Language Processing, NLP). Начнем с основ:

### Работа с символами

#### 1. Доступ к символам в строке
Символы в строке можно обращаться по индексу, как в списках:

```python
text = "Hello, World!"
print(text[0])  # Выведет: 'H'
print(text[-1])  # Выведет: '!'
```

#### 2. Проверка свойств символов
Функции `isalpha()`, `isdigit()`, `isspace()` и другие помогают проверять свойства символов:

```python
text = "Hello, World!"
print(text.isalpha())   # Проверка, состоит ли строка только из букв
print(text.isdigit())   # Проверка, состоит ли строка только из цифр
print(text.isspace())   # Проверка, состоит ли строка только из пробельных символов
```

### Работа со строками

#### 1. Слияние строк
Слияние строк можно выполнить с помощью оператора `+` или метода `join()`:

```python
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print(result)  # Выведет: "Hello World"

# Используя метод join():
words = ["Hello", "World"]
result = " ".join(words)
print(result)  # Выведет: "Hello World"
```

#### 2. Разделение строк на подстроки
Метод `split()` разделяет строку на подстроки по разделителю (по умолчанию - пробел):

```python
text = "Hello, World!"
words = text.split(", ")  # Разделение по запятой с пробелом
print(words)  # Выведет: ['Hello', 'World!']
```

#### 3. Форматирование строк
Форматирование строк можно делать с помощью метода `format()` или f-строк (f-strings):

```python
name = "Alice"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)  # Выведет: "My name is Alice and I am 30 years old."

# Используя f-строки:
message = f"My name is {name} and I am {age} years old."
print(message)  # Выведет то же самое
```

### Текстовая обработка в NLP

#### 1. Преобразование регистра
Методы `lower()` и `upper()` позволяют изменить регистр строки:

```python
text = "Hello, World!"
lower_text = text.lower()
upper_text = text.upper()
print(lower_text)  # Выведет: "hello, world!"
print(upper_text)  # Выведет: "HELLO, WORLD!"
```

#### 2. Удаление пробельных символов
Методы `strip()`, `lstrip()` и `rstrip()` удаляют пробельные символы с начала и конца строки:

```python
text = "   Hello, World!   "
stripped_text = text.strip()
print(stripped_text)  # Выведет: "Hello, World!"
```

#### 3. Замена подстроки
Метод `replace()` заменяет одну подстроку на другую:

```python
text = "Hello, World!"
new_text = text.replace("Hello", "Hi")
print(new_text)  # Выведет: "Hi, World!"
```

### Работа с подстроками

#### 1. Поиск подстроки
Методы `find()` и `index()` используются для поиска подстроки в строке:

```python
text = "Hello, World!"
print(text.find("World"))   # Возвращает индекс начала подстроки (если не найдено, возвращает -1)
print(text.index("World"))  # Возвращает индекс начала подстроки (если не найдено, вызывает ValueError)
```

#### 2. Проверка начала и конца строки
Методы `startswith()` и `endswith()` проверяют, начинается ли или заканчивается ли строка определенной подстрокой:

```python
text = "Hello, World!"
print(text.startswith("Hello"))  # Проверка, начинается ли строка с "Hello"
print(text.endswith("!"))        # Проверка, заканчивается ли строка на "!"
```

### Манипуляции с символами

#### 1. Разделение строки на символы
Используя цикл или генераторы списков, можно разделить строку на отдельные символы:

```python
text = "Hello"
characters = list(text)  # Разделение строки на список символов
print(characters)        # Выведет: ['H', 'e', 'l', 'l', 'o']
```

#### 2. Изменение отдельных символов
Так как строки в Python неизменяемы, для замены отдельного символа требуется создать новую строку:

```python
text = "Hello"
new_text = text[:3] + 'p' + text[4:]  # Замена символа 'l' на 'p'
print(new_text)  # Выведет: "Helpo"
```

### Регулярные выражения

#### 1. Поиск и замена с использованием регулярных выражений
Модуль `re` позволяет выполнять более сложные операции поиска и замены в строках с помощью регулярных выражений:

```python
import re

text = "Hello, World! Today is 2024-07-03."
pattern = r'\d{4}-\d{2}-\d{2}'  # Шаблон для поиска даты в формате YYYY-MM-DD

dates = re.findall(pattern, text)  # Находит все даты в тексте
print(dates)  # Выведет: ['2024-07-03']

new_text = re.sub(pattern, 'DATE', text)  # Заменяет найденные даты на строку 'DATE'
print(new_text)  # Выведет: "Hello, World! Today is DATE."
```

### Обработка символов Unicode

#### 1. Работа с Unicode в Python
Python хорошо поддерживает Unicode, что важно для обработки текстов на различных языках и символов:

```python
text = "你好，世界!"
print(len(text))  # Выведет: 6 (количество символов, включая китайские иероглифы)
```

#### 2. Кодирование и декодирование строк
Для работы с различными кодировками можно использовать методы `encode()` и `decode()`:

```python
text = "Hello, World!"
encoded_text = text.encode('utf-8')  # Кодирование строки в UTF-8
decoded_text = encoded_text.decode('utf-8')  # Декодирование строки из UTF-8
```

#### 3. Срезы строк
Срезы позволяют получать подстроки из строки. Синтаксис среза: `text[start:end:step]`, где `start` - индекс начала (включительно), `end` - индекс конца (не включается), `step` - шаг (по умолчанию 1).

```python
text = "Hello, World!"

# Получение подстроки с 2 по 5 символы (индексы 2, 3, 4)
substring = text[2:5]
print(substring)  # Выведет: 'llo'

# Получение каждого второго символа
every_second = text[::2]
print(every_second)  # Выведет: 'HloWrd'
```


##### 4. Длина строки
Длину строки можно получить с помощью функции `len()`:

```python
text = "Hello, World!"
print(len(text))  # Выведет: 13 (общее количество символов в строке)
```
